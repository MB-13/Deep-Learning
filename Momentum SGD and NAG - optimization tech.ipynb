```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_circles
```

```python
X,y = make_circles(n_samples=200,factor=0.2,noise=0.3)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
```

```python
model = Sequential()
model.add(Dense(10,input_dim=2,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
sgd_m = SGD(learning_rate=0.01,momentum=0.9,nesterov=False)
```

```python
model.compile(loss='binary_crossentropy',optimizer=sgd_m,metrics=['accuracy'])
history = model.fit(X,y,validation_split=0.2,epochs=1000)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X,y.astype('int'),clf=model,legend=2)
```

```python
# Nesterov Accelerated Gradient(NAG)
```

```python
modeln = Sequential()
modeln.add(Dense(10,input_dim=2,activation='relu'))
modeln.add(Dense(10,activation='relu'))
modeln.add(Dense(1,activation='sigmoid'))
```

```python
sgd_nag = SGD(learning_rate=0.01,momentum=0.9,nesterov=True)
```

```python
modeln.compile(loss='binary_crossentropy',optimizer=sgd_nag,metrics=['accuracy'])
historyn = modeln.fit(X,y,validation_split=0.2,epochs=1000)
```

```python
plt.plot(historyn.history['loss'])
plt.plot(historyn.history['val_loss'])
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X,y.astype('int'),clf=modeln,legend=2)
```
