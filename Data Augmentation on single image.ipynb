```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('cat_1/cat.10000.jpg',target_size=(200,200))
```

```python
img
```

```python
type(img)
```

```python
img = image.img_to_array(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python
datagen = ImageDataGenerator(
    rotation_range = 30,
    shear_range = 0.2,
    zoom_range= 0.2,
    horizontal_flip = True,
    width_shift_range = 0.2,
    height_shift_range = 0.2
    )
```

```python
i=0

for output in datagen.flow(input_batch,batch_size=1,save_to_dir='cat_1'):
    i=i+1
    
    if i==20:
        break
```
