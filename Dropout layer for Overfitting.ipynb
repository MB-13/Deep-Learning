```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# REGRESSION DATA GENERATION
```

```python
x_train = np.linspace(-1,1,20)
```

```python
y_train = np.array([-0.6561,-0.3099,-0.59035,-0.50855,-0.285,-0.2443,-0.02445,0.00135,-0.2006,0.07475,-0.1422,0.06515,0.15265,0.3521,0.28415,0.5524,0.23115,0.20835,0.4211,0.60485])
```

```python
x_test = np.linspace(-1,1,20)
```

```python
y_test = np.array([-0.69415,-0.451,-0.43005,-0.4484,-0.1475,-0.5019,-0.28055,0.24595,-0.21425,-0.0286,0.23415,0.46575,0.07955,0.1973,0.0719,0.3639,0.5536,0.3365,0.50705,0.33435])
```

```python
plt.scatter(x_train,y_train,c='black',label='train')
plt.scatter(x_test,y_test,c='red',label='test')
plt.legend()
plt.show()
```

```python
model = Sequential()

model.add(Dense(128,input_dim=1,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(1,activation='linear'))
adam = Adam(learning_rate=0.01)
model.compile(loss='mse',optimizer=adam,metrics=['mse'])
history = model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=500,verbose=1)
```

```python
_, train_mse = model.evaluate(x_train,y_train)
_,test_mse = model.evaluate(x_test,y_test)
print("train: {} test: {}".format(train_mse,test_mse))
```

```python
y_pred = model.predict(x_test)
```

```python
plt.scatter(x_train,y_train,c='black',label='train')
plt.scatter(x_test,y_test,c='red',label='test')
plt.plot(x_test,y_pred)
plt.legend()
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

Overfitting happening here

```python
#     using dropout in regression neural network    #
```

```python
model = Sequential()

model.add(Dense(128,input_dim=1,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1,activation='linear'))
adam = Adam(learning_rate=0.01)
model.compile(loss='mse',optimizer=adam, metrics = ['mse'])
history = model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=500)
```

```python
model.summary()
```

```python
_, train_mse = model.evaluate(x_train,y_train)
_,test_mse = model.evaluate(x_test,y_test)
print("train: {} test: {}".format(train_mse,test_mse))
```

```python
y_pred = model.predict(x_test)
```

```python
plt.scatter(x_train,y_train,c='black',label='train')
plt.scatter(x_test,y_test,c='red',label='test')
plt.plot(x_test,y_pred)
plt.legend()
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

Overfitting reduced by adding dropout layer---> basically it drops some nodes randomly from a layer for each epochs

# Dropout in classification

```python
from sklearn.datasets import make_classification 
```

```python
X,y = make_classification(n_samples= 100,n_features = 2, n_informative=1, n_redundant=0,n_classes=2, n_clusters_per_class=1,class_sep=0.91)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
model = Sequential()

model.add(Dense(128,input_dim=2,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

adam = Adam(learning_rate = 0.01)
model.compile(loss='binary_crossentropy',optimizer = adam,metrics=['accuracy'])
history = model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=500)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X,y.astype('int'),clf=model,legend=2)
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```

```python
# using dropout layer
```

```python
model = Sequential()

model.add(Dense(128,input_dim=2,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1,activation='sigmoid'))

adam = Adam(learning_rate=0.2)
model.compile(loss='binary_crossentropy',optimizer=adam,metrics=['accuracy'])
history = model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=500)
```

```python
plot_decision_regions(X,y.astype('int'),clf=model,legend=2)
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```

# Tips 

1-> overfitting- increase p(that is dropout value)
    underfitting- decrease p
    
2-> first apply dropout layer only before last layer then see results and proceed further if no much changed

3-> CNN - p = 40-50%      
    RNN - p = 20-30%     
    ANN - p = 10-50%
